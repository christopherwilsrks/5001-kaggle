```python
import time
import urllib
from datetime import datetime
import json
import re
from random import random
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from dateutil.relativedelta import relativedelta
```

```python
def get_raw_html(day_str):
    header = {
        "referer": f"https://darksky.net/details/22.282,114.1578/{day_str}/si12/en",
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36 Edg/85.0.564.63',
        'sec-fetch-dest': 'document',
        'sec-fetch-mode': 'navigate',
        'sec-fetch-site': 'same-origin',
        'sec-fetch-user': '?1',
        'upgrade-insecure-requests': '1'
    }
    url = f'https://darksky.net/details/22.282,114.1578/{day_str}/si12/en'
    try:
        r = requests.get(url, header)
    except urllib.error.HTTPError as e:
        print (e.code)
        print (e.reaoson)
        exit()
    except urllib.error.URLError as e:
        print(e)
        exit()
    except:
        exit()
    return r.text
```

```python
def parse_html(raw_text):
    soup = BeautifulSoup(raw_text, 'html.parser')
    pattern = re.compile(r"var hours = \[.*]", re.DOTALL)
    script = soup.find('script', text=pattern)
    pattern = re.compile(r"\[.*]", re.DOTALL)
    begin_idx = script.string.find('[')
    end_idx = script.string.find(']')
    return json.loads(script.string[begin_idx:end_idx+1])
```

```python
def deal_with_json(hours):
    for hour in hours:
        # 解析时间戳
        timestamp = datetime.fromtimestamp(hour['time'])
        hour['time'] = timestamp.strftime('%d/%m/%Y %H:%M')
        hour['date'] = hour.pop('time')
        if not 'summary' in hour:
            hour['summary'] = 'Clear'
    return hours
```

```python
def get_day_weather(day_str):
    hours = deal_with_json(parse_html(get_raw_html(day_str)))
    df = pd.DataFrame(columns=['date','summary','precipIntensity','precipProbability','temperature','apparentTemperature','dewPoint','humidity','windSpeed','windGust','windBearing','uvIndex','visibility'])
    df = df.append(hours)
    # df = df.dropna(axis=1, how='any')
    if 'cloudCover' in df.columns:
        df = df.drop(['cloudCover'], 1)
    if 'icon' in df.columns:
        df = df.drop(['icon'], 1)
    df = df.loc[:, ['date','summary','precipIntensity','precipProbability','temperature','apparentTemperature','dewPoint','humidity','windSpeed','windGust','windBearing','uvIndex','visibility']]
    return df
```

```python
previous_df = pd.read_csv('weather.csv', ',')
if len(previous_df) == 0:
    previous_finish_date = datetime(2016, 12, 31)
else:
    previous_finish_date = datetime.strptime(previous_df.iloc[-1, 0].split()[0], '%d/%m/%Y')
begin_date = previous_finish_date + relativedelta(days=1)
day = begin_date
end_date = datetime(2019, 1, 1)

while day < end_date:

    day_str = day.strftime('%Y-%m-%d')

    df = get_day_weather(day_str)
    df.to_csv('weather.csv', ',', header=False, index=False, mode='a+')
    print(f'write {day_str} to file.')

    day += relativedelta(days=1)
    time.sleep(random()*4)
```
